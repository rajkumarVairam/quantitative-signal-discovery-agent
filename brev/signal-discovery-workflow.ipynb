<h1 id="understanding-workflow">1) Understanding Signal Discovery</h1>

<h2 id="what-is-signal-discovery">1.1) What is Signal Discovery?</h2>

Signal discovery is the process of finding quantitative signals that have predictive power for future stock returns. In quantitative finance, signals are numerical characteristics of stocks that can be used to:

- **Predict returns**: Signals with high Information Coefficient (IC) can help forecast which stocks will outperform
- **Construct portfolios**: Long stocks with favorable signal values, short those with unfavorable values
- **Manage risk**: Understand exposures to different sources of returns

Traditional signal research is labor-intensive, requiring researchers to:
1. Formulate hypotheses about what might predict returns
2. Implement the signal calculation in code
3. Backtest the signal's performance
4. Iterate based on results

This workflow automates that loop with LLM-assisted signal generation and code generation.

<h2 id="workflow-architecture">1.2) Workflow Architecture</h2>

![Workflow Architecture](../notebooks/images/workflow-architecture.png)

<div style="background-color: #f3e5f5; border-left: 6px solid #9c27b0; padding: 15px; margin: 15px 0;">
<h4 style="margin-top: 0;">💡 Closed-Loop Optimization</h4>
The workflow is orchestrated by the <strong>NeMo Agent Toolkit</strong> and uses a closed-loop optimization approach:
<ol>
<li><strong>Signal Agent</strong> receives price-volume data and operators, then generates signal expressions using the LLM</li>
<li><strong>Code Agent</strong> converts signal expressions into executable Python code and includes the required operator implementations from <code>calculator.json</code></li>
<li><strong>Eval Agent</strong> evaluates the signal's predictive power via backtesting (Rank IC)
<ul>
<li>If IC meets threshold → Accept and produce <strong>backtesting results</strong></li>
<li>If IC is poor → Generate <strong>optimization suggestions</strong> and feed them back to the Signal Agent for a new iteration</li>
</ul>
</li>
</ol>
</div>

<h2 id="analyzing-signals">5.2) Analyzing Generated Signals</h2>

The cell below automatically loads the most recent signal output and displays the generated signal expressions, which signal was selected for evaluation, and the corresponding IC metrics.

````python
import json
import re
from pathlib import Path

def _match_signal_name(func_name: str, display_name: str) -> bool:
    """Check if a Python function name corresponds to a signal's display name."""
    func_tokens = set(func_name.lower().replace("signal_", "").split("_"))
    display_tokens = set(re.split(r"[\s\-_]+", display_name.lower()))
    return func_tokens.issubset(display_tokens)

def _parse_signal_expressions(signal_description: str) -> list:
    """Extract the signal JSON array from the LLM's signal_description output."""
    match = re.search(r"```json\s*(\[.*?\])\s*```", signal_description, re.DOTALL)
    if match:
        try:
            return json.loads(match.group(1))
        except json.JSONDecodeError:
            pass
    try:
        return json.loads(signal_description)
    except (json.JSONDecodeError, TypeError):
        return []

def interpret_ic_results(result_path: str):
    """
    Interpret the IC evaluation results from a saved signal JSON file.
    """
    with open(result_path, "r", encoding="utf-8") as f:
        result = json.load(f)
    
    print("=" * 60)
    print("SIGNAL EVALUATION RESULTS")
    print("=" * 60)
    
    print(f"\nTimestamp: {result.get('timestamp', 'N/A')}")
    print(f"Iteration: {result.get('iteration', 'N/A')}")
    
    selected = result.get("selected_signal", "")
    
    signals = _parse_signal_expressions(result.get("signal_description", ""))
    if signals:
        print(f"\nGenerated Signals ({len(signals)}):")
        print("-" * 40)
        for i, signal in enumerate(signals, 1):
            name = signal.get('name', 'N/A')
            is_selected = selected and _match_signal_name(selected, name)
            marker = " << SELECTED" if is_selected else ""
            print(f"\n  Signal {i}: {name}{marker}")
            print(f"    Formula:    {signal.get('formula', 'N/A')}")
            print(f"    Meaning:    {signal.get('meaning', 'N/A')}")
            print(f"    Category:   {signal.get('category', 'N/A')}")
            print(f"    Data fields: {', '.join(signal.get('data_fields_used', []))}")
            print(f"    Operators:   {', '.join(signal.get('operators_used', []))}")
            print(f"    Lookback:    {signal.get('lookback_periods', 'N/A')}")
    
    metrics = result.get("evaluation_metrics", {})
    if metrics:
        metric_label = f"Evaluation Metrics (on: {selected})" if selected else "Evaluation Metrics"
        print(f"\n{metric_label}:")
        print("-" * 40)
        print(f"  Mean IC: {metrics.get('mean_ic', 'N/A'):.4f}" if metrics.get('mean_ic') is not None else "  Mean IC: N/A")
        print(f"  IC Std: {metrics.get('ic_std', 'N/A'):.4f}" if metrics.get('ic_std') is not None else "  IC Std: N/A")
        print(f"  IC IR: {metrics.get('ic_ir', 'N/A'):.4f}" if metrics.get('ic_ir') is not None else "  IC IR: N/A")
        print(f"  T-stat: {metrics.get('t_stat', 'N/A'):.4f}" if metrics.get('t_stat') is not None else "  T-stat: N/A")
        print(f"  P-value: {metrics.get('p_value', 'N/A'):.6f}" if metrics.get('p_value') is not None else "  P-value: N/A")
        print(f"  Num Periods: {metrics.get('num_periods', 'N/A')}")
        print(f"  Positive IC Ratio: {metrics.get('positive_ic_ratio', 'N/A'):.2%}" if metrics.get('positive_ic_ratio') is not None else "  Positive IC Ratio: N/A")
        
        percentiles = metrics.get("ic_percentiles", {})
        if percentiles:
            print("\n  IC Percentiles:")
            for pct, val in percentiles.items():
                print(f"    {pct}: {val:.4f}")
    
    signal_code = result.get("signal_code", "")
    if signal_code:
        print("\nSignal Code:")
        print("-" * 40)
        print(signal_code)

# Automatically pick the most recent output file
output_dir = Path("../src/signal_discovery_workflow/output")
signal_files = sorted(output_dir.glob("signal_*.json"), key=lambda f: f.stat().st_mtime, reverse=True)
if signal_files:
    print(f"Showing results for: {signal_files[0].name}\n")
    interpret_ic_results(str(signal_files[0]))
else:
    print("No signal output files found. Run the workflow first!")
````

This means the optimization loop stops after at most <strong>two iterations</strong> — if the generated signal does not meet the IC threshold on the first attempt, the workflow collects optimization feedback and tries once more before accepting the best result found so far (a "best effort" outcome).

More iterations give the LLM additional chances to refine its signal formulas based on evaluation feedback, which often leads to stronger signals.
